```python
import pandas as pd


def get_unique_values_for_all_columns(csv_file_path):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(csv_file_path)

    # Iterate through each column and print unique values
    unique_values_dict = {}
    for column in df.columns:
        unique_values_dict[column] = df[column].unique()

    return unique_values_dict


# Example usage
# Replace with the path to your CSV file
csv_file_path = '/data_link/servilla/DR_HGNN/dtinew.csv'

unique_values_dict = get_unique_values_for_all_columns(csv_file_path)

# Print unique values for each column
for column, unique_values in unique_values_dict.items():
    print(f"Unique values in column '{column}':")
    print(unique_values)
    print("\n")
```

```python
import pandas as pd


def get_unique_values_and_counts(csv_file_path):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(csv_file_path)

    # Dictionary to store unique values and their counts
    unique_values_counts_dict = {}

    # Iterate through each column and get unique values and their counts
    for column in df.columns:
        unique_values = df[column].unique()
        unique_count = len(unique_values)
        unique_values_counts_dict[column] = {
            "unique_values": unique_values,
            "count": unique_count
        }

    return unique_values_counts_dict


# Example usage
# Replace with the path to your CSV file
csv_file_path = '/data_link/servilla/DR_HGNN/dtinew.csv'

unique_values_counts_dict = get_unique_values_and_counts(csv_file_path)

# Print unique values and counts for each column
for column, unique_data in unique_values_counts_dict.items():
    print(f"Column '{column}':")
    print(f"Total unique values count: {unique_data['count']}")
    print("Unique values:")
    print(unique_data['unique_values'])
    print("\n")
```
